function with outputs and 2f means how many digit will do after index point.

```python
def calculate_gain(initial_price, final_price):
  gain=(final_price-initial_price)/ initial_price
  return gain

result =calculate_gain(100,200)
print(f' total percentage gain: {result * 100:.2f}%')
```
