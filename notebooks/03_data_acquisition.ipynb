```python
import pandas as pd
import glob
import os

# 1. Finds all file names
all_filenames = glob.glob(os.path.join('.', '*.csv'))

# 2. Reads and stacks all files into one DataFrame
master_df = pd.concat(
    [pd.read_csv(f) for f in all_filenames],
    ignore_index=True
)
```
