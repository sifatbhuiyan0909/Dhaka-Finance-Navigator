```python
# 2. MERGE, RENAME, & CONVERT (Run this cell NOW)
import pandas as pd
import glob
import os

# --- MERGE ---
all_filenames = glob.glob(os.path.join('.', '*.csv'))
master_df = pd.concat([pd.read_csv(f, header=None) for f in all_filenames], ignore_index=True)

# --- RENAME & CONVERT ---
master_df.columns = ['Ticker', 'Date', 'Open', 'High', 'Low', 'Close', 'Volume']
master_df['Date'] = pd.to_datetime(master_df['Date'], format='%d-%m-%Y')
```

```python
# CODE FOR DAY 4: Setting the Index
# 1. We create a working copy (df) of the master data.
df = master_df.copy()

# 2. Set the 'Date' column as the DataFrame index
df = df.set_index('Date')

# 3. Check the first few rows to confirm the Date column is now the index
print("--- DataFrame Head after Setting Index ---")
print(df.head())
```

```python
# CORRECTED CODE: We add [0] to select the column before using .unique()
non_numeric_volume = df['Volume'].astype(str).str.extract(r'([^0-9\.\,])').dropna()[0].unique()

print("Non-numeric characters found in Volume:")
print(non_numeric_volume)
```

```python
# 1. Replace the problematic dash/hyphen with an empty string
df['Volume'] = df['Volume'].astype(str).str.replace('-', '', regex=False)

# 2. Convert to numeric, forcing any remaining bad data to NaN (which we fix tomorrow!)
df['Volume'] = pd.to_numeric(df['Volume'], errors='coerce')

# 3. Check the dtypes again to confirm
print("--- DataFrame Info after Volume Cleaning ---")
df.info()
```

```python
# CODE TO CLEAN REMAINING OBJECT COLUMNS: Open and Low
for col in ['Open', 'Low']:
    # 1. Replace the dash/hyphen with an empty string
    df[col] = df[col].astype(str).str.replace('-', '', regex=False)

    # 2. Convert to numeric, forcing any remaining bad data to NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("--- Final DataFrame Info for Day 4 ---")
df.info()
```
